```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10,7)) 

d=np.genfromtxt('NANOGrav_sensitivity.csv')
f=d[:,0]
Omega=d[:,1]
plt.fill_between(f,Omega,1,linewidth=0,color='black',alpha=0.1)
plt.plot(f,Omega,linewidth=8,color='black',alpha=0.5)
plt.text(6e-9,1e-7,'NANOGrav',fontsize=16,color='black',alpha=0.5,rotation=90)

d=np.genfromtxt('SKA_sensitivity.csv')
f=d[:,0]
Omega=d[:,1]
plt.fill_between(f,Omega,1,linewidth=0,color='blue',alpha=0.1)
plt.plot(f,Omega,linewidth=4,color='blue',linestyle='solid')
plt.text(2e-9,1e-7,'PTA(SKA)',fontsize=16,color='blue',alpha=0.5,rotation=90)

d=np.genfromtxt('LISA_sensitivity_2019.csv')
f=d[:,0]
Omega=d[:,1]
plt.fill_between(f,Omega,1,linewidth=0,color='blue',alpha=0.1)
plt.plot(f,Omega,linewidth=4,color='blue')
plt.text(3e-5,8e-6,'LISA',fontsize=16,color='blue',alpha=0.5,rotation=90)

x=[4e-9,1e-8,1e-7,1e-7,1e-8, 4e-9]
y=[4e-9,8e-9,0.5e-6,0.8e-12,0.4e-10,2e-10]
plt.fill(x,y,color='orange',alpha=0.7)


x=[4e-9,1e-8,1e-7,1e-7,1e-8, 4e-9]
y=[4e-9,5e-8,1e-4,3e-8,1e-9,0.5e-10]
plt.fill(x,y,color='green',alpha=0.3)


plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$\tilde f$, Hz',fontsize=16)
plt.ylabel(r'$d\Omega_{GW}/d$log$f$',fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.ylim(1e-15,1e-4)
plt.xlim(0.5e-9,3e-4)
plt.savefig('NANOGrav_GW.png',format='png',dpi=150)

```
